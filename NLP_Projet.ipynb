```python
import re
import nltk
from nltk.corpus import stopwords

# Télécharger les stopwords pour la première fois
nltk.download('stopwords')

# Stopwords en arabe (peuvent être personnalisés)
stop_words_nltk = set(stopwords.words('arabic'))
stop_words_custom = [
    'إلى', 'في', 'من', 'على', 'أن', 'عن', 'ما', 'مع', 'كان', 'أو', 'لكن', 'لأن', 'أي', 'هذا', 'و', 'الذي', 'التي',
    'لها', 'بين', 'عندما', 'أمام', 'خلال', 'منذ', 'لذلك', 'غير', 'كل', 'يجب', 'لن', 'يكون', 'عليه', 'فقط',
    'هذه', 'تلك', 'هل', 'إن', 'لكن', 'ذلك', 'أين', 'كيف', 'متى'
]
stop_words = stop_words_nltk.union(stop_words_custom)

# Liste de phrases sans importance à supprimer
irrelevant_phrases = [
    "المقالة تحتاج إلى مصادر إضافية",
    "من ويكيبيديا، الموسوعة الحرة",
    "هذا النص يمكن أن يكون مشكلاً",
    "حقوق النشر محفوظة"
]

# Fonctions de nettoyage
def remove_foreign_words(text):
    """Supprime les mots contenant des caractères non arabes."""
    return re.sub(r'\b[^\u0600-\u06FF]+\b', ' ', text)

def remove_diacritics(text):
    """Supprime les diacritiques en arabe."""
    return re.sub(r'[\u064B-\u065F]', '', text)

def remove_punctuation_and_symbols(text):
    """Supprime la ponctuation et les symboles non pertinents."""
    return re.sub(r'[،.؟!:\-(){}\[\];"\'~@#$%^&*_+]', '', text)

def remove_numbers_and_symbols(text):
    """Supprime les chiffres et autres symboles."""
    return re.sub(r'[0-9\u0660-\u0669\u06F0-\u06F9%$#@!&^]', '', text)

def remove_multiple_spaces(text):
    """Raccourcit les espaces multiples."""
    return re.sub(r'\s+', ' ', text).strip()

def remove_irrelevant_phrases(text):
    """Supprime les phrases non pertinentes."""
    for phrase in irrelevant_phrases:
        text = re.sub(phrase, '', text)
    return remove_multiple_spaces(text)

def remove_stopwords(text):
    """Supprime les stopwords."""
    words = text.split()
    return ' '.join(word for word in words if word not in stop_words)

# Fonction principale de nettoyage
def clean_text(text):
    """Applique toutes les étapes de nettoyage."""
    text = remove_foreign_words(text)
    text = remove_diacritics(text)
    text = remove_punctuation_and_symbols(text)
    text = remove_numbers_and_symbols(text)
    text = remove_irrelevant_phrases(text)
    return remove_stopwords(text)

# Exemple de texte
text = """
التعليم هو أساس تقدم الأمم وازدهارها. يعتبر التعليم حقًا من حقوق الإنسان الأساسية، وهو أداة حيوية لتحقيق التنمية الشاملة. يوفر التعليم الفرص للأفراد لتطوير مهاراتهم واكتساب المعرفة اللازمة لتحقيق النجاح في الحياة. يعد التعليم في مرحلة الطفولة المبكرة أساسيًا لبناء أسس قوية في حياة الأطفال، بينما يمكن أن يساعد التعليم العالي في تطوير الابتكار وتعزيز الاقتصاد الوطني.
"""

# Nettoyer le texte
cleaned_text = clean_text(text)
print("Texte nettoyé :", cleaned_text)
```

# **Yake :**

```python
# Installer la bibliothèque yake si ce n'est pas déjà fait
# !pip install yake

from yake import KeywordExtractor

# Fonction pour extraire les mots-clés avec YAKE
def extract_with_yake(text, top_n=10, lang="ar"):
    """
    Extrait des mots-clés en utilisant YAKE.
    
    Args:
        text (str): Le texte à analyser.
        top_n (int): Nombre de mots-clés à extraire.
        lang (str): Langue du texte (par défaut 'ar' pour l'arabe).
    
    Returns:
        list: Liste de tuples contenant les mots-clés et leurs scores.
    """
    extractor = KeywordExtractor(lan=lang, n=1, top=top_n)
    keywords_with_scores = extractor.extract_keywords(text)
    
    # Trier par ordre croissant des scores (les plus pertinents en haut)
    sorted_keywords = sorted(keywords_with_scores, key=lambda x: x[1], reverse=False)
    
    return sorted_keywords

# Extraction des mots-clés
mots_cles = extract_with_yake(cleaned_text, top_n=10)

# Afficher les mots-clés et leurs scores
for mot, score in mots_cles:
    print(f"Mot-clé : {mot}, Score : {score}")

# Écrire les résultats dans un fichier texte
with open("mots_cles_yake.txt", "w", encoding="utf-8") as f:
    for mot, score in mots_cles:
        f.write(f"Mot-clé : {mot}, Score : {score}\n")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords

# Télécharger les stopwords en arabe via NLTK
nltk.download('stopwords')
arabic_stopwords = stopwords.words('arabic')


# Liste de textes pour TF-IDF (ici un seul texte mais peut être étendu à plusieurs documents)
documents = [cleaned_text]

# Initialiser le TF-IDF Vectorizer avec les stopwords arabes de NLTK
tfidf_vectorizer = TfidfVectorizer(stop_words=arabic_stopwords, max_features=10)  # Limite à 10 mots-clés

# Appliquer le TF-IDF au texte nettoyé
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Extraire les termes (mots) associés aux scores TF-IDF
terms = tfidf_vectorizer.get_feature_names_out()

# Récupérer les scores TF-IDF pour chaque mot
scores = tfidf_matrix.toarray().flatten()

# Créer un DataFrame pour afficher les mots-clés avec leurs scores
keywords_df = pd.DataFrame({
    'Keyword': terms,
    'TF-IDF Score': scores
})

# Trier par score de TF-IDF en ordre décroissant
keywords_df = keywords_df.sort_values(by='TF-IDF Score', ascending=False)

# Affichage des résultats
print("\nMots-clés extraits avec TF-IDF (et leurs scores) :")
print(keywords_df)
```

# **AraBERT :**

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

# Charger le tokenizer et le modèle AraBERT
model_name = "aubmindlab/bert-base-arabertv02"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForMaskedLM.from_pretrained(model_name)

# Tokenizer le texte
inputs = tokenizer(cleaned_text, return_tensors="pt", truncation=True, padding=True)

# Passer le texte dans le modèle
with torch.no_grad():
    outputs = model(**inputs)

# Obtenir les scores de probabilité des tokens
logits = outputs.logits

# Calculer les scores pour chaque token (utilisation de max sur les logits)
token_scores = []
for token, logit in zip(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]), logits[0]):
    if not token.startswith("##") and not token.endswith("##"):  # Ignorer les sous-tokens
        token_scores.append((token, logit.max().item()))

# Trier les tokens par leur score (du plus élevé au plus bas)
sorted_token_scores = sorted(token_scores, key=lambda x: x[1], reverse=True)

# Supprimer les doublons tout en conservant l'ordre
unique_tokens = {}
for token, score in sorted_token_scores:
    if token not in unique_tokens:  # Ajoute uniquement si le token n'est pas encore enregistré
        unique_tokens[token] = score

# Afficher les 10 mots uniques avec les scores les plus élevés
for token, score in list(unique_tokens.items())[:10]:
    print(f"{token}: {score}")
```

# **AraBERT avec tfidf :**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import AutoTokenizer, AutoModel
import torch
import re
from nltk.corpus import stopwords
import nltk

# Assurez-vous d'avoir les ressources nécessaires

# Charger AraBERT
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabert")
model = AutoModel.from_pretrained("aubmindlab/bert-base-arabert")

# Calcul du score TF-IDF
def calculate_tfidf(corpus):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = vectorizer.get_feature_names_out()
    scores = tfidf_matrix.toarray().flatten()
    return feature_names, scores

# Calcul du score AraBERT
def calculate_arabert_score(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze().detach()  # Embedding de la [CLS] token
    return cls_embedding.mean().item()  # Retourne un score moyen comme un indicateur simple

# Étape 1 : Score TF-IDF
feature_names, tfidf_scores = calculate_tfidf([cleaned_text])

# Étape 2 : Score AraBERT pour chaque mot-clé TF-IDF
arabert_scores = {word: calculate_arabert_score(word) for word in feature_names}

# Étape 3 : Combinaison des scores
final_scores = {word: tfidf_scores[i] + arabert_scores[word] for i, word in enumerate(feature_names)}

# Trier les mots-clés par score
sorted_keywords = sorted(final_scores.items(), key=lambda x: x[1], reverse=True)

top_keywords = sorted_keywords[:5]

for word, score in top_keywords:
    print(f"{word}: {score:.4f}")
```

# **TF-IDF & Yake & AraBERT :**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from yake import KeywordExtractor  # Utilisation de YAKE pour l'extraction des mots-clés
from transformers import AutoTokenizer, AutoModel
import torch
import pandas as pd

# Initialisation du modèle AraBERT
tokenizer_arabert = AutoTokenizer.from_pretrained('aubmindlab/bert-base-arabert')
model_arabert = AutoModel.from_pretrained('aubmindlab/bert-base-arabert')

def calculate_tfidf(corpus, top_n=10):
    """Applique le modèle TF-IDF au texte et retourne les mots-clés et leurs scores."""
    vectorizer = TfidfVectorizer(stop_words=arabic_stopwords, max_features=top_n)
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = vectorizer.get_feature_names_out()
    scores = tfidf_matrix.toarray().flatten()
    
    # Trier les mots-clés par score décroissant
    keywords_scores = list(zip(feature_names, scores))
    sorted_keywords_scores = sorted(keywords_scores, key=lambda x: x[1], reverse=True)
    
    return sorted_keywords_scores

def calculate_arabert_scores(words):
    """Calcule les scores de AraBERT pour chaque mot donné."""
    scores = {}
    for word in words:
        inputs = tokenizer_arabert(word, return_tensors="pt", truncation=True, padding=True, max_length=512)
        with torch.no_grad():
            outputs = model_arabert(**inputs)
        cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze()  # Embedding de la token [CLS]
        scores[word] = cls_embedding.mean().item()  # Retourne la moyenne des valeurs de l'embedding
    return scores

def calculate_yake(text):
    """Utilise YAKE pour extraire les mots-clés."""
    extractor = KeywordExtractor()
    keywords = extractor.extract_keywords(text)
    return [(keyword, score) for keyword, score in keywords]

def extract_with_tfidf_yake_arabert(text, top_n=10):
    """Extrait les mots-clés avec une combinaison de TF-IDF, YAKE et AraBERT."""
    # Nettoyage du texte
    cleaned_text = clean_text(text)
    
    # Extraction des mots-clés avec TF-IDF
    tfidf_keywords = calculate_tfidf([cleaned_text], top_n)
    
    # Extraction des scores AraBERT pour chaque mot-clé
    arabert_scores = calculate_arabert_scores([word for word, _ in tfidf_keywords])
    
    # Extraction des mots-clés avec YAKE
    yake_keywords = calculate_yake(cleaned_text)
    
    # Combinaison des scores TF-IDF, YAKE et AraBERT
    combined_scores = {}
    for word, tfidf_score in tfidf_keywords:
        yake_score = next((score for keyword, score in yake_keywords if word in keyword), 0)
        arabert_score = arabert_scores.get(word, 0)
        combined_scores[word] = tfidf_score + yake_score + arabert_score
    
    # Trier les mots-clés par score combiné
    sorted_keywords = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_keywords[:top_n]

top_keywords = extract_with_tfidf_yake_arabert(cleaned_text, top_n=10)

# Affichage des résultats
top_keywords_df = pd.DataFrame(top_keywords, columns=["Mot-Clé", "Score"])
top_keywords_df
```
